## Data Aquisition
As this is an example and not an actual experiment data was generated using simplified equations for demonstration purposes. One set of data was designed to model a year of growth outside with seasonal variations. The second set was designed to model a traditional laboratory growth experiment. For both datasets the models were multiplied by and random 80% to 120% factor to add some variability.

[**Weather Data**](Example_HourlyWeatherData.xlsx)

The biomass productivity (g m<sup>-2</sup> day<sup>-1</sup>) in both datasets was modeled using weekly averages for the weather data and the Photosynthetic Photon Flux. The equation used is below as reference. Because it a simply equation the calculated values will make examples of statistical analysis simplified compared to real world data. Below is the equation:

$$ Light \ Dur \ (min) * 0.0015 + PAR \ (µmol \ m^{-2} \ s^{-1})*0.01+Air \ Temp \ (°C)*0.008+Water \ Temp \ (°C)*0.03 = Biomass \ Productivity \ (g \ m^{-2} \ day m^{-1}) $$

*Note that this is simplified example and your actual models will not be equations with simple linear components. Actual biomass productivity is also much lower.*

### Time-Series Growth
* **Weather Data** - Weather data from a datalogger was simulated using hourly climate normals generated by the National Oceanic and Atmospheric Administration. Datasets can be downloaded at the [Climate Data Online Search](https://www.ncdc.noaa.gov/cdo-web/search) website.
* **Solar Radiation** - The Photosynthetic Photon Flux (µmol m<sup>-2</sup> s<sup>-1</sup>) estimates were generated through web scraping an existing calculator. A further description can be found at the [Apogee Instruments - Clear Sky Calculator](http://www.clearskycalculator.com/) website.

[**Time-Series Growth Data**](Example_TimeSeriesProductivity.xlsx)

### Laboratory Experiment
The laboratory experiment was based on all combinations of a multiple independent variable design as follow:

Light Dur (min)|PAR (µmol m<sup>-2</sup> s<sup>-1</sup>) |Water Temp (°C)
-----|-----|----- 
600|400|15
780|800|20
960|1200|25

This resulted in 27 individual experimental setups. Each was replicated both times.
